```python
# Import library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Buat dummy data
data = {
    "Group": ["Media", "Acting", "Media", "Acting", "Politics", "Military",
              "Science", "Comedy", "Media", "Acting", "Politics"],
    "Year": [2001, 2000, 2003, 2002, 2005, 2007, 2008, 2001, 2008, 2000, 2006]
}
```

```python
# Buat dataframe
df = pd.DataFrame(data)

df.head()
```

```python
# Ubah data menjadi bentuk matriks
data_heatmap = pd.crosstab(df['Group'], df['Year'])

data_heatmap
```

```python
# Buat heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data_heatmap, cmap='magma', linewidth=0.5, linecolor='gray')

plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title("Distribusi Jumlah Group per Tahun")

plt.tight_layout()
plt.show()
```
