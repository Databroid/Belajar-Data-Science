# 📥 Data Collection (Beginner Level)
Pada notebook ini, kita akan belajar cara dasar mengumpulkan data dari website menggunakan teknik *web scraping* dengan Python. Kita akan menggunakan `requests` untuk mengambil data dari halaman web dan `BeautifulSoup` untuk parsing HTML.


```python
# Install library yang dibutuhkan (jika belum terpasang)
# !pip install requests beautifulsoup4
```

```python
import requests
from bs4 import BeautifulSoup
```

## 1. Ambil halaman web
Kita akan menggunakan halaman contoh dari books.toscrape.com untuk latihan.

```python
url = 'http://books.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.title.text)
```

## 2. Ambil data judul buku
Kita akan ambil semua judul buku dari halaman pertama.

```python
titles = soup.find_all('h3')
for title in titles:
    print(title.a['title'])
```
